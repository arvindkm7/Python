```python
class Item:
    def __init__(self, name : str, price :float, quantity : int): # Python passes Instance as first argument
        self.name = name
        print(f"An instance created : {name}")
        self.price = price
        self.quantity = quantity
        
    def calculate_total_price(self):
        return self.price * self.quantity

item_1 = Item('phone', 10000, 5)
item_2 = Item('mobile', 1000, 500)
print(item_1.calculate_total_price())  
print(item_2.calculate_total_price())
```
