> ## Object Oreinted Programming
>- __Class__ : Template for creating objects. All objects created using the same class will have the same characteristics.
>- __Object__ : An Instance of a Class.
>- __Instantiate__ : Create an instance of a Class.
>- __Method__ : A function defined in a Class.
>- __Attribute__ : A variable bounds to an instance of Class.

```python
class car(object):
    def __init__(self, brand : str, fuel_type : str) -> None:
        self.brand = brand
        self.fuel_type = fuel_type 
        
    def drive(self, distance : int) -> None:
        print(f"Driving {self.brand} for {distance}km [{self.fuel_type}]")
        
        
# Volvo is the instance of class CAR | volvo is the object of type CAR
volvo : car = car(brand = "VOLVO", fuel_type = "Petrol") 
bmw : car = car(brand = 'BMW', fuel_type = 'Electric')

# It is simlar to the models that is called in Machine Learning
print(volvo.brand)
print(volvo.fuel_type)
print(volvo.drive(distance = 50))
```

```python
# Python3 program to display Prime numbers till N

#function to check if a given number is prime
def isPrime(n):
  #since 0 and 1 is not prime return false.
    if(n==1 or n==0):  return False
  
  #Run a loop from 2 to n-1
    for i in range(2,n):
    #if the number is divisible by i, then n is not a prime number.
        if(n%i==0):
            return False
  
  #otherwise, n is prime number.
    return True



# Driver code
N = 100;
#check for every number from 1 to N
for i in range(1,N+1):
  #check if current number is prime
    if(isPrime(i)):
        print(i,end=" ")
```
