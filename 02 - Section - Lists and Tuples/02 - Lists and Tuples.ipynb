## ID
> - The ID for an object may be different each time you run the program but ehile your program's running ,the object will have the same ID.<br>
> - If python has to destroy the object and re create it, then its ID will change.<br>
> - REFER : ID(Object) - docs.python.org

```python
# Boolean is immutable

result = True
another_result = result
print(id(result))
print(id(another_result))

result = False
print(id(result))
# the ID's are  changed which indicates that the Bool is immutable
```

## Mutable Objects
>- The Value can be changed :
    - List
    - Dict
    - Set
    - Bytearray
>- List is Mutable - they can be changed. When we appended a new item, Python was able to change the contents of the list, without creating a new one
    
## Immutable Objects
>- String, Bool are Immutable. When we tried ti change a string. Python created a new Object - a new String, Bool and re-attached the name to it
    

```python
computer_part = ["computer",
                "monitor",
                "keyboard",
                "mouse",
                "mouse mat"]

print(f"Unchanged List\nID : {id(computer_part)}\nList : {computer_part}")
computer_part += ['speaker']
# computer_part.append('speaker')
print(f"\nChanged List\nID : {id(computer_part)}\nList : {computer_part}")

# the ID's are not changed which indicates that the list is mutable
```

### Methods and Functions
>  - __METHOD__ : Method is same as a funtion, except that it's bound to an object. That means we need an object, in order to call the Method.<br>
> - __FUNTIONS__ : When you calla function, you just type its name, and provide any arguments in  parantheses. For example : len(even) gave us the number of items in the list called even.<br> 
> - When we call a method, we tell it which object it's called on. In other words, which object it should be using when it performs its function. For example : the entry we're looking ar, in the table is s.append(x).
***

### Dot Notation

> - __Syntax__ : Start with theobject you're using, then a dot, then the name of the method.<br>
> - If method needs arguments, you put them in parentheses after the method name.<br>
> - For example : __s.append(x)__ <br>
    -  __s__ : The method that we're calling the method on. In this case, we append to a sequence called s <br>
    - __.append__ : The method we want to call. Here, we call the append method<br>
    - __(x)__ : The Argument for the method. Value of x will be appended to the list.

### Case-Insensitive Sorting

```python
sorted_alphabets = sorted('The Quick brown fox jumps over the lazy dog', key = str.casefold)
print(sorted_alphabets)
```

### TUPLES
> - Tuple are another sequence type, along with string, lists and range.
> - Tuples are __IMMUTABLE__
> - Tuples can be __indexed__
> - Advantages <br>
    - Using a tuple, for things that shouldn't change, can help to prevent bugs in the program.<br>
    - Tuple can always unpack successfully, you will always know how many items there are to unpack.

```python
t = "a", 'b', "c"
print(t)

# Verifying 

print(type(t))

# Indexing  in tuple

print(t[0])
print(t[1])
print(t[2]) 

# Unpacking of tuples

a = b = c = d = e = 6
print(c)

x, y = 5, 6
print(f"x : {x}")
print(f"y : {y}")
 
data = 1, 2, 3  # data represent a Tuple
a, b, c = data
print(f"a : {a}")
print(f"b : {b}")
print(f"c : {c}")
print(type(data))
print(type(a))
```
