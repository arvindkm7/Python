```python
import time
import random
from time import time as my_timer

input('Press enter to start ')

wait_time = random.randint(1, 6)
time.sleep(wait_time)
start_time = my_timer()

input('Press enter to stop')

end_time = my_timer()

print('Started at : '+ time.strftime("%X", time.localtime(start_time)))
print('Ended at : ' + time.strftime("%X", time.localtime(end_time)))
print(f'Your reaction  time was {end_time - start_time}')
```

> ## Funtions

```python
def center_text(text):
    left_margin = (80 - len(text)) // 2
    print(" " * left_margin, text)

# Call the Funtion
center_text("Spam and eggs")
center_text("Spam, Spam and eggs")
center_text("Spam, Spam, Spam and eggs")
```

```python
def center_text(*args):
    text = ""
    for arg in args:
        text += str(arg) + " "
    left_margin = (80 - len(text)) // 2
    print(" " * left_margin, text)

# Call the Funtion
center_text("first", "second", "third", 20, 30)
```

```python
def center_text(*args, sep = ':', end = '\n', center = None, flush= False):
    text = ""
    for arg in args:
        text += str(arg) + sep
    left_margin = (80 - len(text)) // 2
    print(" " * left_margin, text, end = end, file = center, flush = flush)

# Call the Funtion
center_text("first", "second", "third", 20, 30)
```

```python
def center_text(text):
    left_margin = (80 - len(text)) // 2
    print(" " * left_margin, text)

# Call the Funtion
center_text("Spam and eggs")
center_text("Spam, Spam and eggs")
center_text("Spam, Spam, Spam and eggs")
```

```python
def center_text(*args, sep = ' ',end = '\n', file = None, flush = False):
    text = ''
    for arg in args:
        text += str(arg) + sep
    left_margin = (80 - len(text)) // 2
    print(" " * left_margin, text)
    return " " * left_margin, text

    
with open('centered',mode = 'w') as centered_file:
    # Call the Funtion
    center_text("Spam and eggs",file = centered_file)
    center_text("Spam, Spam and eggs",file = centered_file)
    center_text("Spam, Spam, Spam and eggs",file = centered_file)
    center_text("First","second","spam",3,5,sep = ":",file = centered_file)
```
