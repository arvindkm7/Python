## CONTINUE



```python
shopping_list = ['milk', 'pasta', 'eggs', 'spam', "bread", 'rice']

for item in shopping_list :
    if item == 'spam':
        continue
    print("Buy " + item)
```

#### Binary Search

```python
low = 1
high = 1000
search_target = int(input(f"Please give the number between {low} and {high} : "))


while guess != search_target:
    guess = int((high + low) / 2)
    if guess == search_target:
        print(f"Welldone, number guessed is correct : {search_target}")
    elif search_target < guess:
        print(f"Guess Higher : {guess}")
        high = int((high + low) / 2)
        guess = int((high + low) / 2)
        print(f"Guess : {guess}\nLow : {low}\nHigh : {high}")
    else:
        print(f"Guess Lower : {guess}")  
        low = int((high + low) / 2)
        guess = int((high + low) / 2)
        print(f"Guess : {guess}\nLow : {low}\nHigh : {high}")
    print()
```
