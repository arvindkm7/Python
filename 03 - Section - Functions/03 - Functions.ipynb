### Methods
> - A function that's bound to na instance of a class is called a __Method__
> - __print__ is a function, but __.sort__ in __my_list.sort()__ is a method. We have seen some of the string methods, such as __str.casefold()__.
> - You use methods in the same way as you use functions, but specify the object that they will act on, before the dot.

```python
# Funtions_Intro
 
# Function defination
def square(n): 
    result_1 = n * n
    result_2 = n * n * n
    return result_1, result_2


# Calling function
n = int(input('Enter the number to find the square and cube : '))
answer_1, answer_2 = square(n) 
print(f'The square and cube of {n} is {answer_1} and {answer_2} ')
```

### Functions

> - #### __Function Defination__  | def square(n):
    - __def__ 
        - A function defination starts with the keyword __def__.
    - __square__ 
        -  name of the defination.
    - __(n)__ 
        - if the function takes parameters, they are declared in parentheses.
        - The parentheses are required, even when there are no parameters.
> - #### __Funtion Call__ | square(n) 
    - __square__
        -  name of the funtion defination.
    - __n__ | __Parameters__
        - __Parameters__ are values that  are passed to function.
        - __Parameters__ are variables, but they're given a value when you call the function.
        - You may see them referred to as __Formal Parameters__.
    - __Arguments__ 
        - __Arguments__ are the values that will be used as __formal parameters__, when we call function. In this case, the real value are the INPUT by the user.
        - Each parameter must be given a value, by providing an argument in the function call.
        - Providing values as arguments is called __passing__ the arguments.
        - If a funtion defines two parameters, we __pass__ two arguments to it when we call it.
        - Some funtion values can have a __Default__ values.
> - For example, __n__ is the parameters of the __square__ function, while __123__ are the arguments passed to the function during the call.

```python
def palindrome_sentence(sentence):
    words = []
    for alphabets in sentence:
        if alphabets.isalpha():
            words.append(alphabets)
    print(words)
    print(words[::-1])
    return words[::-1] == words

sentence = str(input("Enter the sentence to check if it is palindrome or not : "))
print(palindrome_sentence(sentence.casefold()))

# Input : Do Geese see god ?
```

> ### Fibonacci Number.

```python
def fib(n):
    lst = []
    count = 1
    num = 0
    num_2 = [0, 1]
    while count <= n:
        
        if n <= 2:
            lst.append(0)
            lst.append(1)
        elif n > 2:
            lst.append(0)
            lst.append(1)
            for i in range(1,n):
                n_1 = lst[-1] 
                n_2 = lst[-2]
                n_next = n_1 + n_2 
                lst.append(n_next)
        return lst
n = int(input("Enter the number : "))
print(f"The Fibonacci Series till {n} is : {fib(n)})")
```
